```python
import os
import PIL
from PIL import Image
import glob
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import regularizers
from tensorflow.keras import optimizers
import matplotlib.pyplot as plt
import numpy as np
import time
%matplotlib inline
```

### The 378 x 248 x 3 (RGB) sized contour plot image is downsampled to size 28 x 28 x 3 using bicubic interpolation

```python
batch = 23
val_split =0.2

#load training dataset
print("\nLoading training dataset :")
image_train_ds= tf.keras.preprocessing.image_dataset_from_directory(
    'plot_images_train', labels='inferred', label_mode='int', validation_split=val_split ,seed=123,
  subset="training",color_mode='rgb', batch_size=batch, image_size=(28,28), shuffle=True,interpolation='bicubic')

#load validation dataset
print("\nLoading validation dataset :")
image_val_ds= tf.keras.preprocessing.image_dataset_from_directory(
    'plot_images_train', labels='inferred', label_mode='int', validation_split=val_split,seed=100,
  subset="validation",color_mode='rgb', batch_size=batch, image_size=(28,28), shuffle=True,interpolation='bicubic')
```

```python
#load test data
print("\nLoading test dataset :")
image_test_ds= tf.keras.preprocessing.image_dataset_from_directory(
    'plot_images_test', labels='inferred', label_mode='int',color_mode='rgb', batch_size=batch, image_size=(28,28), shuffle=True,interpolation='bicubic')

#load new images for prediction
print("\nLoading new images for prediction :")
image_pred_ds = tf.keras.preprocessing.image_dataset_from_directory(
    'plot_images_pred', labels='inferred', label_mode='int',color_mode='rgb', batch_size=batch, image_size=(28,28), shuffle=False,interpolation='bicubic')
```

```python
#visualization

import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
for images, labels in image_train_ds.take(1):
  for i in range(9):
    ax = plt.subplot(5, 2, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(f"class label = {labels[i]}")
    plt.axis("off")
```

```python
#clear previous model
keras.backend.clear_session()
```

```python
#CNN2 based on alexnet
# 2 Convolution layers with 'relu' activation, 16 and 32 filters each with kernel size 5
# 2 MaxPooling layers with kernel size 2
# one dense layer 
# output layer has 4 neurons (4 classes)


n_classes  = 4
drop_p = 0.5
lambda_val = 0.02

cnn2_model = tf.keras.Sequential([
  layers.experimental.preprocessing.Normalization(mean = 0,variance = 1), #normalization
  layers.Conv2D(16, 5, activation='relu',padding ='same',activity_regularizer= regularizers.L2(lambda_val)),
  layers.MaxPooling2D(strides=1),
  layers.Conv2D(32, 5, activation='relu',padding ='same',activity_regularizer= regularizers.L2(lambda_val)),
  layers.MaxPooling2D(strides=1),
  layers.Flatten(),
  layers.Dense(1024, activation='relu',activity_regularizer= regularizers.L2(lambda_val)),
  layers.Dropout(drop_p),
  layers.Dense(n_classes,activation='softmax')
])

cnn2_model
```

```python
learn_rate = 0.005
opt = tf.keras.optimizers.Adam(learning_rate=learn_rate)

cnn2_model.compile(
  optimizer=opt,
  loss=tf.losses.SparseCategoricalCrossentropy(from_logits=False),
  metrics=['accuracy'])
```

```python
history = cnn2_model.fit(
  image_train_ds,
  validation_data=image_val_ds,
  epochs = 15
)
```

```python
#list all data in history
print(history.history.keys())

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
from tensorflow.keras import backend as K

print("No. of layers: ", len(cnn2_model.layers))
print("Model summary: ", cnn2_model.summary())
```

```python
#evaluate of test data
eval = cnn2_model.evaluate(image_test_ds)
```

```python
#run on prediction dataset
pred_labels = []
true_labels = []
for images_batch,labels_batch in image_pred_ds:
    #print(len(images_batch),len(labels_batch))
    p_labels = cnn2_model.predict(images_batch)
    pred_labels.append(np.argmax(p_labels,axis = 1))
    true_labels.append(np.array(labels_batch))

true_labels = np.concatenate(true_labels).ravel()
pred_labels = np.concatenate(pred_labels).ravel()
print("original labels :",true_labels)
print("predicted labels :",pred_labels)
```

```python
#confusion matrix
res = tf.math.confusion_matrix(pred_labels,true_labels)
print("confusion matrix:",res)

pred_acc = np.array(tf.math.count_nonzero(pred_labels  == true_labels))/len(true_labels)
print("accuracy:",pred_acc)
```
